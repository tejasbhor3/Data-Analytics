## How to handle two line in single plot


```python
x=['Jan','Feb','March','April','May','June']
y2020=[45,67,22,11,55,44]
y2019=[25,56,23,59,81,55]
plt.plot(x,y2019,'bo-')
plt.plot(x,y2020,'r*--')
plt.xlabel('Month')
plt.ylabel('Rainfall')
plt.title('Monthly Rainfall Analysis')
```

## Add legend() which accept label from plot()

```python
x=['Jan','Feb','March','April','May','June']
y2020=[45,67,22,11,55,44]
y2019=[25,56,23,59,81,55]
plt.plot(x,y2019,'bo-',label='Y 2020')
plt.plot(x,y2020,'r*--',label='Y 2019')
plt.xlabel('Month')
plt.ylabel('Rainfall')
plt.title('Monthly Rainfall Analysis')
plt.legend()
```

```python
plt.figure(figsize=(10,10))

plt.subplot(2,2,1)
x=['Jan','Feb','March','April','May','June']
y=[45,67,22,11,55,44]
plt.plot(x,y,'r*--')
plt.xlabel('Month')
plt.ylabel('Rainfall')
plt.title('Year 2020')

plt.subplot(2,2,2)
x=['Jan','Feb','March','April','May','June']
y =[48,69,27,18,57,48]
plt.plot(x,y,'b*--')
plt.xlabel('Month')
plt.ylabel('Rainfall')
plt.title('Year 2019')

plt.subplot(2,2,3)
x=['Jan','Feb','March','April','May','June']
y=[44,64,28,16,52,43]
plt.plot(x,y,'y*--')
plt.xlabel('Month')
plt.ylabel('Rainfall')
plt.title('Year 2018')


plt.subplot(2,2,4)
x=['Jan','Feb','March','April','May','June']
y=[85,47,42,61,95,34]
plt.plot(x,y,'^--')
plt.xlabel('Month')
plt.ylabel('Rainfall')
plt.title('Year 2017')
```

# Bar chart

A bar chart or graph is a chart or graph that presents categorical data with rectangular bars with heights or lengts proportional to the values that they represent. The bars can be plotted certically or horizontally . A vertical bar chart is sometimes called a line graph.

plt.bar()

parameter:

x:[,,,,] list type variable 1 D type
y:[,,,,] list type variable 1 D type
tick_label=rename the coloumn[list type] Ex['Ram,'Shyam'....]
width=0.2,0.4,0.6.....1 thickness of column
color = collection of alternate colors. ex,'r','g','b'


```python
plt.figure(figsize=(10,5))
x=['Jan','Feb','March','April','May','June']
y=[4,6,2,1,5,9]
plt.bar(x,y)
plt.xlabel('Month name')
plt.ylabel('Temp in C')
plt.title("Temperature Analysis")
```

# pie chart
 function: plt.pie()

```python
plt.figure(figsize=(5,5))
y=[56,45,78,15]
plt.pie(y)
plt.show()
```

```python
#display the label
plt.figure(figsize=(5,5))
mylabel=['Apple','Mango','Coconut','Cherry']
y=[56,45,78,15]
plt.pie(y,labels=mylabel)
plt.show()
```

```python
#rotate the start angle
plt.figure(figsize=(5,5))
mylabel=['Apple','Mango','Coconut','Cherry']
y=[56,45,78,15]
plt.pie(y,labels=mylabel,startangle=90)
plt.show()
```

```python
# explode : to put attention on some slice
plt.figure(figsize=(5,5))
mylabel=['Apple','Mango','Coconut','Cherry']
y=[56,45,78,15]
myexplode=[0.1,0,0,0]
plt.pie(y,labels=mylabel,explode=myexplode)
plt.show()
```

```python
# shadow
plt.figure(figsize=(5,5))
mylabel=['Apple','Mango','Coconut','Cherry']
y=[56,45,78,15]
myexplode=[0.1,0,0,0]
plt.pie(y,labels=mylabel,explode=myexplode,shadow=True)
plt.show() 
```

```python
#legend
plt.figure(figsize=(15,10))
mylabel=['Apple','Mango','Coconut','Cherry']
y=[56,45,78,15]
myexplode=[0.1,0,0,0]
plt.pie(y,labels=mylabel,explode=myexplode,shadow=True)
plt.legend()
plt.show() 
```

```python
#percentage is displayed by attribute : autopct
plt.figure(figsize=(8,8))
mylabel=['Apple','Mango','Coconut','Cherry']
y=[56,45,70,50]
myexplode=[0,0.1,0,0]
plt.pie(y,labels=mylabel,explode=myexplode,shadow=True,autopct='%1.2f%%')
plt.legend()
plt.show() 
```
