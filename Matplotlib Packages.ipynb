Pie chart is used to represent percentage distribution of data and is applicable only for categorical variable 

```python
labels='Frog','Hogs','Dogs','Logs'
size=[15,30,45,10]
colors=['yellowgreen','gold','lightskyblue','lightcoral']
explode=(0,0.1,0,0)
plt.pie(size,autopct='%1.1f%%',shadow=True,colors=colors,startangle=90,explode=explode,labels=labels)
plt.axis('equal');
```

```python
N=5
Menmeans=(20,35,30,35,27)
Womenmeans=(25,32,34,20,25)
Menstd=(2,3,4,1,2)
Womenstd=(3,5,2,3,3)
ind=np.arange(N)
width=0.35
P1=plt.bar(ind,Menmeans,width,yerr=Menstd)
P2=plt.bar(ind,Womenmeans,width,bottom=Menmeans,yerr=Womenstd)
plt.ylabel('Scores')
plt.title('Score by Group and Gender')
plt.xticks(ind,('G1','G2','G3','G4','G5'))
plt.yticks(np.arange(0,81,10))
plt.legend((P1[0],P2[0]),('Men','Women'))
plt.show();
```

```python
ind=np.arange(len(Menmeans))
width=0.35
fig,ax=plt.subplots()
rects1=ax.bar(ind-width/2,Menmeans,width,yerr=Menstd,color='SkyBlue',label='Men')
rects2=ax.bar(ind+width/2,Womenmeans,width,yerr=Womenstd,color='IndianRed',label='Women')
ax.set_ylabel('Scores')
ax.set_title('Scores by Group and Gender')
ax.set_xticks(ind)
ax.set_xticklabels(('G1','G2','G3','G4','G5'))
ax.legend();
```

Create an array ranging from 12 to 36, reshape it to 4*6 matrix. Try to reverse the row , reverse the column and reverse both and print them.

```python
a1=np.arange(12,36)
print('a1 array')
print(a1)
a2=a1.reshape(4,6)
print('a2 array')
print(a2)
a3=a2[::-1]
print('a3 array')
print(a3)
a4=a2[::1,::-1]
print('a4 array')
print(a4)
a5=a2[::-1,::-1]
print('a5 array')
print(a5)
```

Create an array ranging from 12 to 36, reshape it to 4*6 matrix. Convert all the values of the matrix to 0 other than the ones in border.

```python
c1=np.arange(12,36)
print('c1 array')
print(c1)
c2=c1.reshape(4,6)
print('c2 array before modification')
print(c2)
c2[1:-1,1:-1]=0
print('c2 array after modification')
print(c2)
```
